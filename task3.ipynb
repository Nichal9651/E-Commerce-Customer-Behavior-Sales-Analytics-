Exploratory Data Analysis (EDA) was performed to understand the structure, patterns, relationships, and anomalies in the e-commerce dataset using statistical summaries and visualizations.

**Data Exploration:**

```python
import pandas as pd
df=pd.read_csv("/content/ecommerce_customer_behavior_dataset.csv")
df.head()
```

***Summary Statistics:***

```python
df.head()
df.info()
df.shape
numeric_cols = df.select_dtypes(include=['int64', 'float64'])
numeric_cols.describe()

```

***Data Visualization:***

a) Histogram â€“ Distribution Analysis

```python
import matplotlib.pyplot as plt

numeric_cols.hist(figsize=(14, 10), bins=20)
plt.suptitle("Distribution of Numerical Variables")
plt.show()
```

***b) Box Plot â€“ Outlier Visualization***

```python
import seaborn as sns

plt.figure(figsize=(8,4))
sns.boxplot(x=df['Total_Amount'])
plt.title("Box Plot of Total Amount")
plt.show()
```

**c)Scatter plot-feature relationship**

```python
plt.figure(figsize=(6,4))
plt.scatter(df['Quantity'], df['Total_Amount'])
plt.xlabel("Quantity")
plt.ylabel("Total Amount")
plt.title("Quantity vs Total Amount")
plt.show()
```

**Correlation Matrix**

```python
plt.figure(figsize=(10,8))
correlation_matrix = numeric_cols.corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

Interpretation:

Strong positive correlation between Quantity and Total_Amount

Moderate relationship between Unit_Price and Total_Amount

Customer engagement metrics show weaker correlations with sales
