***Task 1: Analyze a Dataset's Basic Statistics***







**Dataset Selection:**

```python
import pandas as pd

# Load dataset
df = pd.read_csv("/content/ecommerce_customer_behavior_dataset.csv")

# Quick overview
df.head()
```

Dataset Info


```python
df.info()
```

Describing the dataset

```python
df.describe()
```

**Summary Statistics:**

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64'])

summary_stats = pd.DataFrame({
    "Mean": numeric_cols.mean(),
    "Median": numeric_cols.median(),
    "Standard Deviation": numeric_cols.std()
})

summary_stats
```

**Data Distribution:**

```python
import matplotlib.pyplot as plt

numeric_cols.hist(bins=20, figsize=(14, 10))
plt.suptitle("Distribution of Numerical Variables", fontsize=15)
plt.show()
```

**Visualization :**

```python
mean_values = numeric_cols.mean()
plt.figure(figsize=(12, 5))
mean_values.plot(kind='bar')
plt.title("Mean Values of Numerical Variables")
plt.ylabel("Mean Value")
plt.xlabel("Variables")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Descriptive Analysis:**

Observations:

The Total_Amount and Unit_Price variables show higher variability compared to quantity, indicating differences in product pricing and order value.


The mean values of Total_Amount are higher than their medians, suggesting a right-skewed distribution influenced by high-value purchases.


Session_Duration_Minutes and Pages_Viewed exhibit moderate spread, reflecting varied customer engagement levels.


Customer_Rating shows relatively low variance, indicating generally consistent customer satisfaction.


No missing values were observed in the dataset, improving the reliability of statistical analysis.
